```python
df = df.join(pd.get_dummies(df.gender,dtype='int'))
df = df.join(pd.get_dummies(df.smoking_history,dtype='int'))
```

```python
df.drop(columns=['gender','smoking_history'],inplace=True)
```

```python
df
```

```python
def capping_outlier(df,col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1
    lower_bound = Q1 - 1.5 * Q1
    upper_bound = Q3 + 1.5 * Q3
    df[col] = df[col].clip(lower_bound,upper_bound)
    return df

for col in ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']:
    df = capping_outlier(df,col)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']] = scaler.fit_transform(df[['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']])
```

```python
df.sample(10)
```

```python
target = df.diabetes
target
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(df.drop('diabetes',axis=1),target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models = {
    'Logistic Regression':{
        "model" : LogisticRegression(),
        "params" : {
            'C' : [0.1,1,10],
            'solver' : ['liblinear', 'lbfgs']
        }
    },
    "Random Forest" : {
        'model' : RandomForestClassifier(),
        'params':{
            'n_estimators' : [50,100],
            'max_depth':[None,10,15,20]
        }
    },
    "GaussianNB" : {
        'model' : GaussianNB(),
        'params': {}
    }
    
}
```

```python
from sklearn.model_selection import GridSearchCV

scores1 = []

for model_name , model_par in models.items():
    clf = GridSearchCV(model_par['model'],model_par['params'],cv=5,return_train_score=False)
    clf.fit(xtrain,ytrain)
    scores1.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_param' : clf.best_params_
    })

df55 = pd.DataFrame(scores1,columns=['model','best_score','best_params'])
df55
```

```python
clf = RandomForestClassifier(n_estimators=100,max_depth=10)
clf.fit(xtrain,ytrain)
```

```python
accuracy = clf.score(xtest, ytest)
```
